## transforms 图片处理

```python
import PIL.Image
import matplotlib.pyplot as plt
from torchvision import transforms
import numpy as np
import torch
from torch import nn
```

```python
path_to_image = "data/1.jpg"
# 4通道的 RBGB 转 RGB
rgba_image = PIL.Image.open(path_to_image)
display(np.array(rgba_image).shape)
rgb_image = rgba_image.convert('RGB')
```

```python
imarray = np.asarray(rgb_image) 
# imarray_rgba = np.asarray(rgba_image)   ## 四通道转np
plt.figure()
plt.imshow(imarray)
plt.show()
```

### transforms.RandomResizedCrop

```python
resize_cropper = transforms.RandomResizedCrop(size=(224, 224))
resized_crops = [resize_cropper(rgb_image) if _ != 0 else rgb_image for _ in range(5)]
resized_crops
```

```python
plt.figure(figsize=(20,15))
for i,p in enumerate(resized_crops):
    plt.subplot(1,5,i+1)
    imarray = np.asarray(p) 
    plt.imshow(imarray)  
    
plt.subplots_adjust(hspace=0.35)
plt.show()
```

### transforms.RandomHorizontalFlip

```python
resize_cropper = transforms.RandomHorizontalFlip(0.5)
resized_crops = [resize_cropper(rgb_image) if _ != 0 else rgb_image for _ in range(5)]
resized_crops
```

```python
plt.figure(figsize=(20,15))
for i,p in enumerate(resized_crops):
    plt.subplot(1,5,i+1)
    imarray = np.asarray(p) 
    plt.imshow(imarray)  
    
plt.subplots_adjust(hspace=0.35)
plt.show()
```

### transforms.ToTensor() 和 transforms.Normalize

```python
np.asarray(rgb_image).shape
```

```python
## 原来的图片转换成 numpy 之后 三个维度分别是 [H,W,C] 
np.asarray(rgb_image).shape
```

```python
## 在经过 transforms.ToTensor() 转换之后，维度变成  [C,H,W]，注意直接使用 torch.tensor(numpy_data) 并不会有这中效果
a = np.asarray(rgb_image)
a = transforms.ToTensor()(a)
a.shape
```

```python
Normalize_im = transforms.Normalize([0.485, 0.456, 0.406], [0.229, 0.224, 0.225])(a)
```

```python
plt.figure(figsize=(10,8))
## 原图
plt.subplot(1,3,1)
plt.imshow(imarray)

## 标准化后
plt.subplot(1,3,2)
plt.imshow(Normalize_im[:,:,:].numpy().transpose((1, 2, 0)))

## 标准化后转回来
plt.subplot(1,3,3)
mean = np.array([0.485, 0.456, 0.406])
std = np.array([0.229, 0.224, 0.225])
image = Normalize_im[:,:,:].numpy().transpose((1, 2, 0))    ## 反转或置换数组的轴；返回修改后的数组。    image = std * image + mean
image = std * image + mean
image = np.clip(image, 0, 1)   
plt.imshow(image)

plt.show()
```
