## 垃圾邮件数据分类：

| SPAM E-MAIL DATABASE ATTRIBUTES (in .names format)

| 48 continuous real [0,100] attributes of type word_freq_WORD 
| = percentage of words in the e-mail that match WORD,
| i.e. 100 * (number of times the WORD appears in the e-mail) / 
| total number of words in e-mail.  A "word" in this case is any 
| string of alphanumeric characters bounded by non-alphanumeric 
| characters or end-of-string.

| 6 continuous real [0,100] attributes of type char_freq_CHAR
| = percentage of characters in the e-mail that match CHAR,
| i.e. 100 * (number of CHAR occurences) / total characters in e-mail

| 1 continuous real [1,...] attribute of type capital_run_length_average
| = average length of uninterrupted sequences of capital letters

| 1 continuous integer [1,...] attribute of type capital_run_length_longest
| = length of longest uninterrupted sequence of capital letters

| 1 continuous integer [1,...] attribute of type capital_run_length_total
| = sum of length of uninterrupted sequences of capital letters
| = total number of capital letters in the e-mail

| 1 nominal {0,1} class attribute of type spam
| = denotes whether the e-mail was considered spam (1) or not (0), 
| i.e. unsolicited commercial e-mail.

#### 定义全连接网络


* torch.nn是专门为神经网络设计的模块化接口。nn构建于autograd之上，可以用来定义和运行神经网络。


* nn.Module是nn中十分重要的类,包含网络各层的定义及forward方法。


```python
class DNNclassification(nn.Module):
    def __init__(self):
        super(DNNclassification,self).__init__()
            
        ## 第一个隐藏层
        self.hidden1 = nn.Sequential(
            nn.Linear(
                in_features=57,      ## 第一个隐藏层的输入，即数据的特征数
                out_features=30,     ## 第一个隐藏层的输出，即神经元的数量
                bias=True),
            nn.ReLU()
        )

        ## 第二个隐藏层
        self.hidden2 = nn.Sequential(
            nn.Linear(30,10),
            nn.ReLU()                
        )

        ## 输出层
        self.classification = nn.Sequential(
            nn.Linear(10,2),
            nn.Sigmoid()                
        )

    ## 定义网络的前向传播
    def forward(self,x):
        fc1 = self.hidden1(x)                ## Full Connection 1
        fc2 = self.hidden2(fc1)              ## Full Connection 2
        output = self.classification(fc2)

        return fc1,fc2,output
```

#### 查看网络结构

```python
## 输出 DNN 网络结构
dnnc = DNNclassification()
dnnc
```

## 三、特征工程

#### 标准化

```python
ss = StandardScaler()
X_train_s = ss.fit_transform(X_train)
X_test_s = ss.fit_transform(X_test)
```

## 四、训练模型

#### 4.1、先尝试使用为标准化的数据训练模型

```python
## 将数据转为张量,PyTorch 不接受 aFloatTensor作为分类目标，所以分类标签数据类型要改成LongTensor
X_train_nots = torch.from_numpy(X_train.astype(np.float32))   ## nots 表示 not ss
y_train_t = torch.from_numpy(y_train.astype(np.int64))
X_test_nots = torch.from_numpy(X_test.astype(np.float32))   ## nots 表示 not ss
y_test_t = torch.from_numpy(y_test.astype(np.int64))
```

```python
## 将训练集转为张量后，使用TensorDataset 将x和y打包
train_data_nots = Data.TensorDataset(X_train_nots,y_train_t)
## 定义数据加载器，将训练数据集进行批量处理
train_nots_loader = Data.DataLoader(

    dataset = train_data_nots,       ## 指定使用的数据集
    batch_size= 64 ,                 ## 批处理样本大小
    shuffle = True,                  ## 每次迭代前打乱数据
    num_workers = 1                  ## 开启的工作进程数
)
```

```python
## 先初始化一次dnn 模型
dnnc = DNNclassification()
dnnc
```

#### 训练模型和测试模型

```python
## 设置优化器和损失函数
optimizer = torch.optim.Adam(dnnc.parameters(),lr=0.01)   
loss_func = nn.CrossEntropyLoss()
## 定义训练过程损失指标
history1 = hl.History()
## 使用 Canvas 可视化损失
canvas1 = hl.Canvas()           ## 初始化Canvas 图层对象，用于可视化网络的训练过程
print_steps = 25

for epoch in range(15):
    ## 对训练数据的加载器进行迭代计算
    for step,(b_x,b_y) in enumerate(train_nots_loader): ## DataLoader 指定为batch_size 为64，一共3450条数据，3450/64 ≈ 54 ，迭代54次
        _,_,output = dnnc(b_x)                      ## 定义的 DNN 的输出
        train_loss= loss_func(output,b_y)           ## 使用二分类交叉熵损失函数计算损失
        optimizer.zero_grad()                       ## 每个迭代步的梯度初始化为0
        train_loss.backward()                       ## 损失的反向传播，计算梯度
        optimizer.step()                            ## 使用梯度进行优化，更新网络参数
        niter = epoch*len(train_nots_loader) + step +1  ## 迭代器，用于计数第几次批计算，一次循环有54批，一共15个epoch，一共15*54=810批
        
        ## 计算每经过print_step 次迭代后的输出，25批计算一次
        if niter % print_steps == 0:
            _,_,output = dnnc(X_test_nots)
            _,pre_lab = torch.max(output,1)
            test_accuracy = accuracy_score(y_test_t,pre_lab)
            
            ## 为history添加epoch，损失和精度
            history1.log(niter,train_loss=train_loss,test_accuracy=test_accuracy) ## 添加要可视化的变量，分别为
            ## 使用两个图像可视化损失和精度 -- 可视化网络训练过程
            with canvas1:
                canvas1.draw_plot(history1['train_loss'])     ## 损失尽可能低的更好
                canvas1.draw_plot(history1['test_accuracy'])  ## 验证集的精度尽可能高的更好              
```

可以发现，此时损失函数很难迭代稳定

#### 4.2、接下来使用标准化后的数据来进行训练

```python
## 先初始化一次模型，防止被受之前训练的影响
dnnc = DNNclassification()
dnnc
```

```python
## 将数据转为张量
X_train_s = torch.from_numpy(X_train_s.astype(np.float32))   ## nots 表示 not ss
y_train_t = torch.from_numpy(y_train.astype(np.int64))
X_test_s = torch.from_numpy(X_test_s.astype(np.float32))   ## nots 表示 not ss
y_test_t = torch.from_numpy(y_test.astype(np.int64))
```

```python
## 将训练集转为张量后，使用TensorDataset 将x和y打包
train_data = Data.TensorDataset(X_train_nots,y_train_t)
## 定义数据加载器，将训练数据集进行批量处理
train_loader = Data.DataLoader(

    dataset = train_data,       ## 指定使用的数据集
    batch_size= 64 ,                 ## 批处理样本大小
    shuffle = True,                  ## 每次迭代前打乱数据
    num_workers = 1
)
```

#### 训练模型和测试模型

```python
## 优化器
optimizer = torch.optim.Adam(dnnc.parameters(),lr=0.01)
loss_func = nn.CrossEntropyLoss()
## 定义训练过程损失指标
history1 = hl.History()
## 使用 Canvas 可视化损失
canvas1 = hl.Canvas()
print_steps = 25

for epoch in range(15):
    ## 对训练数据的加载器进行迭代计算
    for step,(b_x,b_y) in enumerate(train_loader): ## 每一次迭代会使用 DataLoader 指定的批处理大小的数据
        _,_,output = dnnc(b_x)                      ## DNN 的输出
        train_loss= loss_func(output,b_y)           ## 使用二分类交叉熵损失函数计算损失
        optimizer.zero_grad()                       ## 初始化梯度为0
        train_loss.backward()                       ## 损失的反向传播，计算梯度
        optimizer.step()                            ## 使用梯度进行优化，更新网络参数
        niter = epoch*len(train_loader) + step +1
        
        ## 计算每经过print_step 次迭代后的输出
        if niter % print_steps == 0:
            _,_,output = dnnc(X_test_s)
            _,pre_lab = torch.max(output,1)
            test_accuracy = accuracy_score(y_test_t,pre_lab)
            
            ## 为history添加epoch，损失和精度
            history1.log(niter,train_loss=train_loss,test_accuracy=test_accuracy)
            ## 使用两个图像可视化损失和精度
            with canvas1:
                canvas1.draw_plot(history1['train_loss'])     ## 损失尽可能低的更好
                canvas1.draw_plot(history1['test_accuracy'])  ## 验证集的精度尽可能高的更好       
```

## 五、TensorBoardx 可视化

#### 参数：

* add_scalar: (tag: str,scalar_value: Union[float, numpy_compatible],global_step: Optional[int] = None,...)

    需要注意：这里的 scalar_value 一定是 float 类型，如果是 PyTorch scalar tensor，则需要调用 .item() 方法获取其数值。
    
    我们一般会使用 add_scalar 方法来记录训练过程的 loss、accuracy、learning rate 等数值的变化，直观地监控训练过程。
    
    ??SummaryWriter.add_scalar

#### tensorboard 选项：

* smoothing ：指数移动平均线，移动平均通常与时间序列数据一起使用，以消除短期波动，突出长期趋势或周期

```python
from tensorboardX import SummaryWriter
```

```python
## 先初始化一次模型，防止被受之前训练的影响
dnnc = DNNclassification()
dnnc
```

```python
SumWriter = SummaryWriter(log_dir='./runs/log4')
## 优化器
optimizer = torch.optim.Adam(dnnc.parameters(),lr=0.01)
loss_func = nn.CrossEntropyLoss()
## 定义训练过程损失指标
history1 = hl.History()
## 使用 Canvas 可视化损失
canvas1 = hl.Canvas()
print_steps = 25

for epoch in range(15):
    ## 对训练数据的加载器进行迭代计算
    for step,(b_x,b_y) in enumerate(train_loader): ## 每一次迭代会使用 DataLoader 指定的批处理大小的数据
        _,_,output = dnnc(b_x)                      ## DNN 的输出
        train_loss= loss_func(output,b_y)           ## 使用二分类交叉熵损失函数计算损失
        optimizer.zero_grad()                       ## 初始化梯度为0
        train_loss.backward()                       ## 损失的反向传播，计算梯度
        optimizer.step()                            ## 使用梯度进行优化，更新网络参数
        niter = epoch*len(train_loader) + step +1
        
        ## 计算每经过print_step 次迭代后的输出
        if niter % print_steps == 0:
            ## 为日志添加训练集损失函数
            SumWriter.add_scalar('train loss',train_loss.item(),global_step = niter)
            _,_,output = dnnc(X_test_s)
            _,pre_lab = torch.max(output,1)
            test_accuracy = accuracy_score(y_test_t,pre_lab)
            ## 为日志添加在测试集上的预测精度
            SumWriter.add_scalar('test acc',test_accuracy.item(),niter)
            
            ## 使用直方图可視化网络中参数的分布情况
            for name,param in dnnc.named_parameters():
                SumWriter.add_histogram(name,param.data.numpy(),niter)
                
SumWriter.add_graph(dnnc,(X_train_s,))
SumWriter.close()  ## 不 close ，数据不会保存
```
