```python
import pandas as pd
import numpy as np

pd.set_option('display.max_columns', None)
```

#### 获取列名

```python
lst = []
with open('./spambase.names','r') as f:
    lst.extend(f.readlines()[33:])
    
columns=[]
for i in lst:
    columns.append(i.split(':')[0])
columns.append('label')
columns
```

#### 读数据

```python
data = pd.read_csv('./spambase.data')
data.head()
```

#### 重新命名列名

```python
a = pd.DataFrame(data.columns).T
a.columns=columns
a

data.columns=columns
```

#### 合并数据

```python
df = pd.concat([a,data],axis=0,ignore_index=True)
df.head()
```

#### 改一个异常值

```python
df.loc[0,'word_freq_all']=0.64
df.loc[0,'word_freq_will']=0.64
df.loc[0,'word_freq_free']=0.64
df.loc[0,'word_freq_cs']=0.64
```

#### 写新文件

```python
df.to_csv('spambase.csv',index=False)
```
