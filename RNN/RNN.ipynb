# RNN


```python
import torch
from torch import nn 
import hiddenlayer as hl
```

```python
rnn = nn.RNN(input_size=4,hidden_size=5,batch_first=True) ## batch_first 表示输入和输出的第一维表示批次的大小
```

```python
rnn
```

```python
inputs = torch.zeros([2, 3, 4])  ## 2个序列，每个序列中长度为3，每个时刻输入大小为4
inputs
```

```python
outputs,hn = rnn(inputs)   ## outputs 为输出序列的隐藏层，hn为最后一个熟客的隐藏层
```

```python
outputs
```

```python
hn
```

#### RNN 网络可视化

```python
## 可视化卷积神经网络
## 输入:[batch, time_step, input_dim]
hl_graph = hl.build_graph(rnn, torch.zeros([2, 3, 4]))
hl_graph.theme = hl.graph.THEMES["blue"].copy()  
hl_graph
```
